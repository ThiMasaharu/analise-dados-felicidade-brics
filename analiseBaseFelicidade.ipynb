```python
# Gráfico 2 - Percepções de Corrupção x Felicidade

# Coluna Perceptions of corruption de cada país

corrupcaoBrasil = dfBrazil.loc[:, "Perceptions of corruption"]
corrupcaoRussia = dfRussia.loc[:, "Perceptions of corruption"]
corrupcaoIndia = dfIndia.loc[:, "Perceptions of corruption"]
corrupcaoChina = dfChina.loc[:, "Perceptions of corruption"]
corrupcaoAfrica = dfAfrica.loc[:, "Perceptions of corruption"]

corrupBrazil = corrupcaoBrasil *10
corrupRussia = corrupcaoRussia *10
corrupIndia = corrupcaoIndia *10
corrupChina = corrupcaoChina *10
corrupAfrica = corrupcaoAfrica *10

# Coluna Avaliação de vida de cada país

felBrazil = dfBrazil.loc[:, "Life Ladder"]
felRussia = dfRussia.loc[:, "Life Ladder"]
felIndia = dfIndia.loc[:, "Life Ladder"]
felChina = dfChina.loc[:, "Life Ladder"]
felAfrica = dfAfrica.loc[:, "Life Ladder"]


# Coluna Anos de cada país

yearBrazil = dfBrazil.loc[:, "year"]
yearRussia = dfRussia.loc[:, "year"]
yearIndia = dfIndia.loc[:, "year"]
yearChina = dfChina.loc[:, "year"]
yearAfrica = dfAfrica.loc[:, "year"]
  
plt.xlabel("Anos")

# Gráfico Brasil

plt.ylabel("Eixo y")
plt.title("Gráfico - Brasil\nPercepção de Corrupção x Felicidade")

plt.scatter(yearBrazil, corrupBrazil, label = "Percepção de corrupção")
plt.scatter(yearBrazil, felBrazil, label = "Felicidade")
# plt.title("Gráfico - China\nPercepção de Corrupção x Felicidade")

# plt.scatter(yearChina, corrupChina, label = "Log PIB per capita")
# plt.scatter(yearChina, felChina, label = "Felicidade")
# plt.plot(yearChina, corrupChina, linestyle = "--")
# plt.plot(yearChina, felChina, linestyle = "--")
# plt.legend()
# plt.show()
plt.plot(yearBrazil, corrupBrazil, linestyle = "--")
plt.plot(yearBrazil, felBrazil, linestyle = "--")
plt.legend()
plt.show()


# Gráfico Russia

plt.ylabel("Eixo y")
plt.title("Gráfico - Russia\nPercepção de Corrupção x Felicidade")

plt.scatter(yearRussia, corrupRussia, label = "Percepção de corrupção")
plt.scatter(yearRussia, felRussia, label = "Felicidade")
plt.plot(yearRussia, corrupRussia, linestyle = "--")
plt.plot(yearRussia, felRussia, linestyle = "--")
plt.legend()
plt.show()


# Gráfico China

plt.ylabel("Eixo y")
plt.title("Gráfico - China\nPercepção de Corrupção x Felicidade")

plt.scatter(yearChina, corrupChina, label = "Percepção de corrupção")
plt.scatter(yearChina, felChina, label = "Felicidade")
plt.plot(yearChina, corrupChina, linestyle = "--")
plt.plot(yearChina, felChina, linestyle = "--")
plt.legend()
plt.show()


# Gráfico India

plt.ylabel("Eixo y")
plt.title("Gráfico - India\nPercepção de Corrupção x Felicidade")

plt.scatter(yearIndia, corrupIndia, label = "Percepção de corrupção")
plt.scatter(yearIndia, felIndia, label = "Felicidade")
plt.plot(yearIndia, corrupIndia, linestyle = "--")
plt.plot(yearIndia, felIndia, linestyle = "--")
plt.legend()
plt.show()


# Gráfico Africa

plt.ylabel("Eixo y")
plt.title("Gráfico - Africa\nPercepção de Corrupção x Felicidade")
plt.ylim(3,10)

plt.scatter(yearAfrica, corrupAfrica, label = "Percepção de corrupção")
plt.scatter(yearAfrica, felAfrica, label = "Felicidade")
plt.plot(yearAfrica, corrupAfrica, linestyle = "--")
plt.plot(yearAfrica, felAfrica, linestyle = "--")
plt.legend()
plt.show()



#   Anotações

# O gráfico da Russia varia poouco, sendo que muitas vezes que o valor de Percepção de corrupção cai, o da felicidade sobe. Sendo assim, o valor do Percepção de corrupção per capta parece não influenciar tanto na felicidade da população do país.
# No gráfico da China, não há dados de percepção de corrupção.
# Na India, o valor de percepção de corrupção varia moderadamente, já o de felicidade cai constantemente. Sendo assim, o valor de corrupção parece não influenciar muito no valor da felicidade.
```

```python
# Gráfico da felicidade BRICS

felBrazil = dfBrazil.loc[:, "Life Ladder"]
felRussia = dfRussia.loc[:, "Life Ladder"]
felIndia = dfIndia.loc[:, "Life Ladder"]
felChina = dfChina.loc[:, "Life Ladder"]
felAfrica = dfAfrica.loc[:, "Life Ladder"]

yearBrazil = dfBrazil.loc[:, "year"]
yearRussia = dfRussia.loc[:, "year"]
yearIndia = dfIndia.loc[:, "year"]
yearChina = dfChina.loc[:, "year"]
yearAfrica = dfAfrica.loc[:, "year"]

felBrazil.mean()


plt.ylabel("Eixo y")
plt.title("Gráfico - BRICS\nFelicidade")

plt.scatter(yearBrazil, felBrazil, label = "Brasil")
plt.scatter(yearRussia, felRussia, label = "Russia")
plt.scatter(yearIndia, felIndia, label = "India")
plt.scatter(yearChina, felChina, label = "China")
plt.scatter(yearAfrica, felAfrica, label = "Africa do Sul")
plt.plot(yearBrazil, felBrazil, linestyle = "--")
plt.plot(yearRussia, felRussia, linestyle = "--")
plt.plot(yearIndia, felIndia, linestyle = "--")
plt.plot(yearChina, felChina, linestyle = "--")
plt.plot(yearAfrica, felAfrica, linestyle = "--")
plt.legend()
plt.show()
```

```python
# Gráfico da média de felicidade BRICS

nomePaisesFelicidade = ["Brasil", "Russia", "India", "China", "África do Sul"]
listaMediaFelicidade = [felBrazil.mean(), felRussia.mean(), felIndia.mean(), felChina.mean(), felAfrica.mean()]


plt.ylabel("Eixo y")
plt.title("Gráfico - BRICS\n Média de Felicidade")

plt.bar(nomePaisesFelicidade, listaMediaFelicidade)
plt.legend()
plt.show()
```
